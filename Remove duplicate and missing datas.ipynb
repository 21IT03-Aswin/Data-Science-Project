```python
import pandas as pd
data={
    'Name':['aswin','aswin','ajay','raja','ajay'],
    'Age':[19,19,18,20,18],
    'city':['vellore','vellore','erode','dharmapuri','erode']
}
df=pd.DataFrame(data)
print("Original data:")
print(df)
```

```python
duplicates=df.duplicated()
print("/n Duplicated rows:")
print(duplicates)
```

```python
duplicate_rows=df[df.duplicated()]
print("/n Duplicated rows:")
print(duplicate_rows)
```

```python
df_no_duplicates = df.drop_duplicates()
print("/n DataFrame after removing duplicates:")
print(df_no_duplicates)
```

```python
df.drop_duplicates(inplace=True)
print("/n DataFrame after removig duplicates  (in place): ")
print(df)
```

```python
df_no_duplicates_groupby=df.groupby('Age').first().reset_index()
print("/n DataFrame after removing duplicates using groupby and first():")
print(df_no_duplicates_groupby)
```

```python
pivot_table=df.pivot_table(index=['Name'],aggfunc='first')
print("/n Pivot table after removing duplicates using pivot_table and first():")
print(pivot_table)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data={
    'Name':['aswin','aswin','ajay','raja','ajay'],
    'Age':[19,19,18,20,18],
    'city':['vellore','vellore','erode','dharmapuri','erode']
}
df=pd.DataFrame(data)
df_unique=df.drop_duplicates()
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.plot(df.index,df['Age'],marker='o',linestyle='-',color='b')
plt.title('Original data')
plt.xlabel('Index')
plt.ylabel('Age')
plt.grid(True)



plt.subplot(1,2,2)
plt.plot(df_unique.index,df_unique['Age'],marker='o',linestyle='--',color='r')
plt.title('Data After Removing Duplicates')
plt.xlabel('Index')
plt.ylabel('Age')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("world_population.csv").head(20)
df_unique=df.drop_duplicates()
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.plot(df.index,df['Rank'],marker='o',linestyle='-',color='b')
plt.title('Original data')
plt.xlabel('Index')
plt.ylabel('Age')
plt.grid(True)



plt.subplot(1,2,2)
plt.plot(df_unique.index,df_unique['Rank'],marker='o',linestyle='--',color='r')
plt.title('Data After Removing Duplicates')
plt.xlabel('Index')
plt.ylabel('Age')
plt.grid(True)

plt.tight_layout()
plt.show()
```
